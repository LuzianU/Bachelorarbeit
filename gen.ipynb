```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# klassifizierten Liniendiagrammen, unterteilt auf 93 mit einer Wertelinie, 284 zusammengesetzte, 94 nicht überlappendene und 336 überlappende Liniendiagramme.

# Define the class labels and the confusion matrix values
labels = [
    "Zusammengesetzt",
    "Nicht überlappend",
    "Überlappend",
    "Eine Linie",
]  # Replace with your actual class labels
cm = np.array(
    [
        [54, 0, 0, 0],
        [1, 12, 2, 0],
        [1, 7, 66, 1],
        [1, 0, 0, 18],
    ]
)  # Replace with your actual confusion matrix values

# Display the confusion matrix with larger fonts
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 24}, xticklabels=labels, yticklabels=labels)
plt.ylabel("Vorhergesagte Klassen", fontsize=20)
plt.xlabel("Wahre Klassen", fontsize=20)
# plt.title("Confusion Matrix", fontsize=24)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# klassifizierten Liniendiagrammen, unterteilt auf 93 mit einer Wertelinie, 284 zusammengesetzte, 94 nicht überlappendene und 336 überlappende Liniendiagramme.

# Define the class labels and the confusion matrix values
labels = [
    "Zusammengesetzt",
    "Mehrere Linien",
    "Eine Linie",
]  # Replace with your actual class labels
cm = np.array(
    [
        [56, 1, 0],
        [1, 85, 1],
        [0, 0, 18],
    ]
)  # Replace with your actual confusion matrix values

# Display the confusion matrix with larger fonts
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", annot_kws={"size": 24}, xticklabels=labels, yticklabels=labels)
plt.ylabel("Vorhergesagte Klassen", fontsize=20)
plt.xlabel("Wahre Klassen", fontsize=20)
# plt.title("Confusion Matrix", fontsize=24)
plt.xticks(fontsize=16, rotation=90)
plt.yticks(fontsize=16, rotation=0)
plt.show()
```
